```python
annotations_file_path, data_dir_path = get_data_path("alip_sample")
```

```python
def show_audio(row):
    audio_path = os.path.join(data_dir_path, row['file_path'])
    waveform, sample_rate = torchaudio.load(audio_path)
    display(f"label: {row['has_hesitation']}", Audio(waveform, rate=sample_rate))

def show_sample_audios_from_filter(n:int=5):
	annotations = pd.read_csv(annotations_file_path)

	annotations[['file_path', 'has_hesitation']].sample(n).apply(show_audio, axis=1)
	print("\n\n\n--------------------------------------------------------\n"*3)
```
